```python
def get_value_or_zero(attribute):
    if attribute is None or attribute == '':
        return 0
    else:
        return attribute.replace(',', '.')

def check_for_ip(attribute):
    if attribute is None or attribute == '':
        return 0
    attribute_str = str(attribute)

    if 'Ip' in attribute_str:
        return 0
    else:
        return attribute_str.replace(',', '.')
```
